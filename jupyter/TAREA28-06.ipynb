```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

datos = pd.read_csv("assets/data/games.csv", index_col="Id")
rating_grafico = datos.copy().sort_values('Rating').head(20)


datos.head(5)
```

```python

print(datos.isnull().any())


print()
filas_con_nulo_datos = datos[datos['Rating'].isnull()]
print(filas_con_nulo_datos.head(13))


print()
num_filas_con_nulo_datos = filas_con_nulo_datos.shape[0]
print('Numero de filas que tienen valor nulo en la columna "Rating":', num_filas_con_nulo_datos)
```

```python

datos['Rating'] = datos['Rating'].fillna(0)


print(datos.isnull().any())
```

```python
rating_menor_mayor_datos = datos.sort_values('Rating')
rating_menor_mayor_datos.head(10)
```

```python

rating_grafico.set_index('Title', inplace=True)

rating_grafico['Rating'].plot(kind='bar')

ax = rating_grafico['Rating'].plot(x=None, kind='bar')

for i, rating in enumerate(rating_grafico['Rating']):
    ax.annotate(str(rating), xy=(i, rating), ha = 'center', va = 'bottom')
```

```python
datos.to_csv('assets/data/games_sin_rating_nulos.csv')
rating_menor_mayor_datos.to_csv('assets/data/games_menor_mayor_rating.csv')
```
